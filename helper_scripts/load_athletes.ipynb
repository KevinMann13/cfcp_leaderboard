```python
import pandas as pd

df = pd.read_excel('~/Downloads/Fall Battle 2021.xlsx', sheet_name="Master Roster")
```

```python
teams = [
    {"id": 1, "name": "Elijah"},
    {"id": 2, "name": "Ka'eo"},
    {"id": 3, "name": "Kulani "},
    {"id": 4, "name": "Oliver"},
    {"id": 5, "name": "BiggDave "},
    {"id": 6, "name": "David G. "},
    {"id": 7, "name": "Genny"},
    {"id": 8, "name": "Russell "},
]

delete_names = ["TJ Moen", "Mark Sells"]
```

```python
compiled = []
id = 1
for t in teams:
    athletes = df[t['name']].dropna().to_list()
    for a in athletes:
        
        skip = False
        for d in delete_names:
            if d in a:
                skip = True
                continue
        if skip:
            continue

        compiled.append((id, t['id'], a))
        id += 1
compiled
```
